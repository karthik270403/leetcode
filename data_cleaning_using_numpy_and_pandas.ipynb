```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
df.columns
```

```python
print(df.age.mean())

df['age'] = df.age.fillna(df.age.mean())
df['age'] = df.age.round(decimals=0)

print(df.age.isnull().sum())
```

```python
df.head()
```

```python
df["established"] = df["established"].replace(-1,np.nan)
df["established"].fillna('Unknown', inplace=True)
```

```python
df.location.astype
```

```python
df['location']= df['location'].str.replace("Australia Aus", "Australia, Aus")
df['location_city'] = df['location'].str.split(",", expand=True).get(0)
df['city_sign'] = df['location'].str.split(",", expand=True).get(1)
df.drop('location', inplace=True, axis=1)
```

```python
df['rating'] = df["rating"].replace(-1,np.nan)
df['rating'] = df["rating"].fillna(df['rating'].mean()).round(decimals=1)
```

```python
df['easy_apply'] = df['easy_apply'].replace("-1", "False")
```

```python
df['salary'].iloc[0]

df['salary'] = df['salary'].apply(lambda x:x.replace("$",''))
df['salary'] = df['salary'].apply(lambda x:x.replace("k",''))
```

```python
# Convert 'item_price' to string before using .str accessor
df["item_price"] = df["item_price"].astype(str).str.replace('[^\d.]', '', regex=True).astype(float)

# Calculate revenue for each item
df['revenue'] = df['quantity'] * df['item_price']

# Group by order_id and calculate total revenue per order
order_grouped = df.groupby(by=['order_id']).sum()

# Calculate the mean revenue per order
mean_revenue = order_grouped['revenue'].mean()
print(mean_revenue)
```
